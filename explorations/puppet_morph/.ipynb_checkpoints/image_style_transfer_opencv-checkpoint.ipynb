# Overall idea
Given two portraits, s and d, we want d to have the same style as of s. For example, if s has a face with open mouth, we want the face in d to have an open mouth too, eventhough the mouth in d is close in the original image. This can be done by deep learning tools and style transfers (see face swap in my other library in GitHub), but here it is done by image morphing.

1) Find the images landmarks (deep learning is used here)
2) Find Delauney triangulation of the faces built on top of the landmarks for d. Essentially, built the triangulation for d, use the same order of landmarks to build the triangles on d for s as well. This makes sure that the triangles point to the same areas as of in d.
3) For each pair of triangles (in d and s, note they are corresponding as they have been built on the same landmark orders), find a transformation from trangle in d to the same triangle in s.

```python
import numpy as np
import cv2
import face_alignment
import face_recognition
import dlib

def add_corners_to_landmarks(landmarks, rect, add_corners=False):
    # Create an array of points as tuples
    landmark_points = []

    for line in landmarks:
        x, y = line[0], line[1]
        landmark_points.append((int(x), int(y)))
    
    # Add corners of the face rect. This is in particular usefulr for style transfer.
    if add_corners:
        lt = (rect[0]+1, rect[1]+1)
        rt = (rect[0]+rect[2]-1, rect[1]+1)
        lb = (rect[0]+1, rect[1]+rect[3]-1)
        rb = (rect[0]+rect[2]-1, rect[1]+rect[3]-1)
        landmark_points.extend([lt, rt, lb, rb]) 
    
    return landmark_points

def get_triangles(image, landmarks, perc_out_face=0.1):

    convexhull = cv2.convexHull(landmarks)
    rect = cv2.boundingRect(convexhull)    
    # Stretch the bounding box a bit to include more of the face. This is not needed if the face swap is the aim.
    # This is useful though if the style is going to transfer.
    if perc_out_face>0.0:    
        w = rect[2]
        h = rect[3]
        t = np.max([rect[1]-np.int32(h*perc_out_face), 0])
        r = np.min([rect[0]+w+np.int32(w*perc_out_face), image.shape[1]-1])
        b = np.min([rect[1]+h+np.int32(h*perc_out_face), image.shape[0]-1])
        l = np.max([rect[0]-np.int32(w*perc_out_face), 0])

        w = r-l
        h = b-t

        rect = (int(l), int(t), int(w), int(h))
    
    subdiv = cv2.Subdiv2D(rect)

    points = add_corners_to_landmarks(landmarks, rect)    
    
    # Now build the Delauney triangles
    subdiv.insert(points)
    triangles = subdiv.getTriangleList()
    triangles = np.array(triangles, dtype=np.int32)

    indexes_triangles = []
    for t in triangles:
        pt1 = (t[0], t[1])
        pt2 = (t[2], t[3])
        pt3 = (t[4], t[5])
        index_pt1 = points.index(pt1)
        index_pt2 = points.index(pt2)
        index_pt3 = points.index(pt3)
        if index_pt1 is not None and index_pt2 is not None and index_pt3 is not None:
            triangle = [index_pt1, index_pt2, index_pt3]
            indexes_triangles.append(triangle)
            
    return indexes_triangles, rect, points

def get_landmarks(img, model='recog'):
    if model is 'align':
        fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._2D, flip_input=False)
        face_landmarks = fa.get_landmarks(img)[0] # Only first face
        
    if model is 'recog':
        face_landmarks_list = face_recognition.face_landmarks(img)
        face_landmarks = []
        for face_component in face_landmarks_list[0].values():
            face_landmarks.extend(list(face_component))    
        face_landmarks = face_landmarks[:-4]
        face_landmarks = np.array(face_landmarks, np.int32)
    
    if model is 'dlib':
        detector = dlib.get_frontal_face_detector()
        predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        face = detector(img_gray)[0]
        landmarks = predictor(img_gray, face)

        face_landmarks = []
        for n in range(0, 68):
            x = landmarks.part(n).x
            y = landmarks.part(n).y
            face_landmarks.append([x, y])
        face_landmarks = np.array(face_landmarks, np.int32)
    
    return face_landmarks

def get_geometry(img, face_landmarks=None, perc_out_face=0.1):
    if face_landmarks is None:
        face_landmarks = get_landmarks(img)
    
    landmarks = np.array(face_landmarks, np.int32)       
    
    triangles, rect, landmarks_points = get_triangles(img, landmarks, perc_out_face=perc_out_face)

    perc_out_face = 0.0

    w = rect[2]
    h = rect[3]
    t = np.max([rect[1]-np.int32(h*perc_out_face), 0])
    r = np.min([rect[0]+w+np.int32(w*perc_out_face), img.shape[1]])
    b = np.min([rect[1]+h+np.int32(h*perc_out_face), img.shape[0]])
    l = np.max([rect[0]-np.int32(w*perc_out_face), 0])

    face = img[t:b, l:r]
        
    # Returns the first face ONLY
    return triangles, rect, landmarks_points, face

image_name = 'open_mouth.jpeg'
# image_name = 'jim_carrey.jpg'
# image_name = 'hillary_clinton.jpg'
img_1 = cv2.imread(''.join([images_path, image_name]))
triangles1, rects1, landmark_points1, face1 = get_geometry(img_1)

# image_name = 'bradley_cooper.jpg'
image_name = 'man.jpg'
# image_name = 'ted_cruz.jpg'
# image_name = 'hillary_clinton.jpg'
img_2 = cv2.imread(''.join([images_path, image_name]))
triangles2, rects2, landmark_points2, face2 = get_geometry(img_2)
```

```python
def show_images(img, triangles, landmark_points, rect):
    img_copy = img.copy()
    for p in landmark_points:
        cv2.circle(img_copy, p, 3, (0, 255, 0), -1)
        
    delaunay_color = (0, 0, 0)

    for t in triangles:
        pt1 = landmark_points[t[0]]
        pt2 = landmark_points[t[1]]
        pt3 = landmark_points[t[2]]
        cv2.line(img_copy, pt1, pt2, delaunay_color, 1)
        cv2.line(img_copy, pt2, pt3, delaunay_color, 1)
        cv2.line(img_copy, pt3, pt1, delaunay_color, 1)

    cv2.imshow('img1', img_copy)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
# show_images(img_2, triangles2, landmark_points2, rects2)
show_images(img_1, triangles2, landmark_points1, rects1)
```

```python
from matplotlib import pyplot as plt

def cropp_triangle(img, landmarks_points, triangle_index):
    # Get triangle vertices. Note, this is in an abstract space, with no pattern.
    tr1_pt1 = landmarks_points[triangle_index[0]]
    tr1_pt2 = landmarks_points[triangle_index[1]]
    tr1_pt3 = landmarks_points[triangle_index[2]]
    triangle = np.array([tr1_pt1, tr1_pt2, tr1_pt3], np.int32)
    
    # Build a rectangle to bound the triangle. Just coordinates, no patterns.
    rect = cv2.boundingRect(triangle)
    (x, y, w, h) = rect
    
    # Crrop that rectangle from the original image, now it has the image pattern in the rectangle.
    # Note that the rectangle is a small part of the image, and its coordinates start from 0 and 0. It 
    # has been crropped from the image and forms a new little image.
    cropped_triangle = img[y: y + h, x: x + w]
    cropped_tr_mask = np.zeros((h, w), np.uint8)
    
    # Now we need to recreate the triangle in the rectangle with the patter, and get rid of patterns outside of the 
    # triangle.
    
    # First, offset points by left top corner of the respective rectangles
    points = np.array([[tr1_pt1[0] - x, tr1_pt1[1] - y],
                       [tr1_pt2[0] - x, tr1_pt2[1] - y],
                       [tr1_pt3[0] - x, tr1_pt3[1] - y]], np.int32)
    
    # Now, mask things out of the triangle.
    cv2.fillConvexPoly(cropped_tr_mask, points, 255)
    
    return cropped_triangle, points, rect, cropped_tr_mask

def transform_triangles(img_s, img_d, landmark_points_s, landmark_points_d, triangles_d):
    # Copies style from image s, fills the image d (if s is laghing, then d would be laghing)
    
    # Triangulation of both faces
    final_mrphed = np.zeros(img_s.shape, dtype = img_s.dtype)

    for triangle_index in triangles_d:
        cropped_triangle_d, points_d, rect_d, cropped_tr_mask_d = cropp_triangle(img_d, landmark_points_d, triangle_index)
        cropped_triangle_s, points_s, rect_s, cropped_tr_mask_s = cropp_triangle(img_s, landmark_points_s, triangle_index)
        
        # The rectangle that contains the traingle to be warpped
        (xs, ys, ws, hs) = rect_s
        
        # Warp triangles
        points_d = np.float32(points_d)
        points_s = np.float32(points_s)
        
        # Find out how to map the triangle in d to the triangle in s (points have no pattern)
        M = cv2.getAffineTransform(points_d, points_s)
        
        # Apply the mapping to the triangle in d with patterns 
        warped_triangle = cv2.warpAffine(cropped_triangle_d, M, (ws, hs))        
        
        # Now remove patterns outside of the triangle after warping
        warped_triangle = cv2.bitwise_and(warped_triangle, warped_triangle, mask=cropped_tr_mask_s)
        
        
        img2_new_face_rect_area = final_mrphed[ys: ys + hs, xs: xs + ws]
        img2_new_face_rect_area_gray = cv2.cvtColor(img2_new_face_rect_area, cv2.COLOR_BGR2GRAY)
        _, mask_triangles_designed = cv2.threshold(img2_new_face_rect_area_gray, 1, 255, cv2.THRESH_BINARY_INV)
        warped_triangle = cv2.bitwise_and(warped_triangle, warped_triangle, mask=mask_triangles_designed)

        img2_new_face_rect_area = cv2.add(img2_new_face_rect_area, warped_triangle)
        final_mrphed[ys: ys + hs, xs: xs + ws] = img2_new_face_rect_area
        
#         img_s_new_face_rect_area = final_mrphed[y_s: y_s + hs, x_s: x_s + ws]
#         img_s_new_face_rect_area = cv2.add(img_s_new_face_rect_area, warped_triangle)
#         final_mrphed[y_s: y_s + hs, x_s: x_s + ws] = img_s_new_face_rect_area
        
    return final_mrphed

# mrp = transform_triangles(img_1, img_2, landmark_points1, landmark_points2, triangles2)
mrp = transform_triangles(img_2, img_1, landmark_points2, landmark_points1, triangles1)
plt.imshow(cv2.cvtColor(mrp, cv2.COLOR_BGR2RGB))
```

```python
img_s = img_2
landmark_points_s = landmark_points2
rect_s = rects2

mask = np.zeros(img_s.shape, dtype = img_s.dtype)  
cnvxhl = cv2.convexHull(np.array(landmark_points_s))
cv2.fillConvexPoly(mask, np.int32(cnvxhl), (255, 255, 255))
r = rect_s
center = ((r[0]+int(r[2]/2), r[1]+int(r[3]/2)))
output = cv2.seamlessClone(np.uint8(mrp), img_s, mask, center, cv2.NORMAL_CLONE)

plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
```
